```python
# dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load dataframe
df = pd.read_csv("C:\\Users\\Chris\\Desktop\\VesselData.csv")
df = df[["eta", "discharge1", "load1", "discharge2", "load2", "discharge3", "load3", "discharge4", "load4", "previousportid", "nextportid"]]
df["eta"] = pd.to_datetime(df["eta"], format="%Y-%m-%d %H:%M:%S.%f")
```

```python
# check if there are any missing values
df.isnull().sum()
```

```python
# Plot the variables against time to have a visual indication of a distribution
df.plot(x="eta", y=["discharge1", "discharge2", "discharge3", "discharge4"], kind="line", figsize=(14,14))
plt.show()
```

```python
df.plot(x="eta", y=["load1", "load2", "load3", "load4"], kind="line", figsize=(14,14))
plt.show()
```
